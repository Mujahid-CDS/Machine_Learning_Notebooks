# SQL WHERE

```python
query = 'SELECT * FROM customers WHERE state="Minnesota"'
res = pd.read_sql(query, conn)
res.head()
```

```python
query = 'SELECT * FROM orders WHERE quantity=3'
res = pd.read_sql(query, conn)
res.head()
```

# SQL JOINS

# left join

```python
query = 'SELECT * FROM orders LEFT JOIN customers ON orders.customer_id=customers.customer_id'
res = pd.read_sql(query, conn)
res.head()
```

```python
query = 'SELECT * FROM orders RIGHT JOIN customers ON orders.customer_id=customers.customer_id'
res = pd.read_sql(query, conn)
res.head()
```

```python
query = 'SELECT * FROM orders LEFT JOIN products ON orders.product_id=products.product_id'
res = pd.read_sql(query, conn)
res.head()
```

# Inner join 

```python
query = 'SELECT * FROM orders INNER JOIN products ON orders.product_id=products.product_id'
res = pd.read_sql(query, conn)
res.head()
```

# Join Multiple Tables

```python
query = 'SELECT * from orders INNER JOIN customers ON orders.customer_id=customers.customer_id INNER JOIN products ON orders.product_id=products.product_id'
res = pd.read_sql(query, conn)
res.head()
```
