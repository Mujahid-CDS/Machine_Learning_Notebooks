```python
import pandas as pd
import numpy as np
```

```python
df1=pd.DataFrame({'name':['sonu','sameera','sasha'],'math':[12,13,14]})
df1
```

```python
df2=pd.DataFrame({'name':['sonu','sameera','sasha'],'eng':[9,8,4]})
df2
```

```python
pd.merge(df1,df2,on='name')
```

```python
print('--------------------')
df3=pd.DataFrame({'name':['sonu','sameera','sasha','basha'],'math':[12,13,14,24]})
print(df3)
print('--------------------')
df4=pd.DataFrame({'name':['sonu','sameera','sasha','darsha'],'eng':[9,8,4,6]})
print(df4)
print('--------------------')
```

```python
pd.merge(df3,df4,on='name') #inner join happens 
```

```python
pd.merge(df3,df4,on='name', how='left') #left join 
```

```python
pd.merge(df3,df4,on='name', how='right') #right join 
```

```python
pd.merge(df3,df4,on='name', how='outer') #outer join
```

```python
df3 = pd.DataFrame({
    'name': ['Arshak', 'Deepak', 'Aiman', 'Sinduja', "Suhail"],
    'eng': [99, 97, 92, 93, 85]
})

df5 = pd.DataFrame({
    'name': ['Sinduja', "Pallavi",'Deepak', 'Aiman','Arshak',  ],
    'math': [99, 97, 92, 93, 96]
})

df6 = pd.DataFrame({
    'name': ['Sinduja', "Pallavi",'Deepak', 'Aiman','Arshak',  'Rudrash'],
    'sci': [95, 99, 95, 93, 95,99]
})

df7=pd.merge(df3,df5,on='name', how='outer')
pd.merge(df7,df6,on='name', how='outer')
```

```python
pd.merge(pd.merge(df3,df5,on='name', how='outer'),df6,on='name', how='outer')
```
