```python
import pandas as pd
import numpy as np
df=pd.read_csv('spam.csv', encoding='latin-1')
```

```python
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
X=df.email
y=df.type
```

```python
X.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vect=TfidfVectorizer()
```

```python
X_tfidf=vect.fit_transform(X)
X_tfidf.toarray()
```

```python
list(X_tfidf.toarray()[0])
```

```python
X_tfidf.toarray().shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test= train_test_split(X_tfidf, y, random_state=50)
```

```python
from sklearn.naive_bayes import BernoulliNB
```

```python
model=BernoulliNB()
```

```python
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report,accuracy_score
```

```python
y_predict=model.predict(X_test)
```

```python
pd.crosstab(y_test,y_predict)
```

```python
accuracy_score(y_test,y_predict)
```

```python
from sklearn.naive_bayes import MultinomialNB
model=MultinomialNB()
model.fit(X_train, y_train)
y_predict=model.predict(X_test)
```

```python
pd.crosstab(y_test,y_predict)
```

```python
accuracy_score(y_test,y_predict)
```

```python
X_train, X_test, y_train, y_test= train_test_split(X_tfidf.toarray(), y, random_state=50)
from sklearn.naive_bayes import GaussianNB
model=GaussianNB()
model.fit(X_train, y_train)
y_predict=model.predict(X_test)
```

```python
pd.crosstab(y_test,y_predict)
```

```python
accuracy_score(y_test,y_predict)
```
