# Visualization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rcParams
```

# Line Plot / Line chart

```python
x=[1,2,3,4,5,6,7,8,9]
y=[23,30,21,12,6,8,17,18,15]
```

```python
plt.plot(x,y)
```

```python
year=[2011,2012,2013,2014,2015,2016,2017,2018,2019]
profit=[100,123,145,156,178,189,203,223,245]
```

```python
plt.plot(year,profit)
```

```python
plt.plot(year,profit)
plt.xlabel('year')
plt.ylabel('profit')
plt.title('year vs profit')
plt.show()
```

# Scatter plot

```python
plt.scatter(year,profit)
plt.xlabel('year')
plt.ylabel('profit')
plt.title('year vs profit')
plt.show()
```

```python
age=[0,5,10,15,20,25]
pocket_money=[0,15,20,30,30,40]
plt.plot(age,pocket_money)
plt.xlabel('age')
plt.ylabel('pocket money')
plt.title('My Awesome Graph')
plt.show()
```

```python
sales=[10,20,30,45,60]
teams=['T1','T2','T3','T4','T5']
```

```python
plt.pie(sales,labels=teams)
plt.title('My Teams sales')
plt.show()
```

```python
profit=[30,50,20,40]
quaters=['q1','q2','q3','q4']
plt.pie(profit,labels=quaters, autopct='%0.3f')
plt.title('My quater sales')
plt.show()
```

```python
profit=[30,50,20,40]
quaters=['q1','q2','q3','q4']
plt.pie(profit,[0.3,0,0.3,0],labels=quaters, autopct='%0.3f')
plt.title('My quater sales')
plt.show()
```
