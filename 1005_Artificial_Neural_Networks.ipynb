```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('breast_cancer.csv')
```

```python
df.describe()
```

```python
X=df.iloc[:,:-1]
y=df.outcome
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
```

```python
scaler.fit(X)
X_scale=pd.DataFrame(scaler.transform(X),columns=X.columns)
```

```python
X_scale.describe()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X_scale, y, test_size=0.2, random_state=35)
```

```python
from sklearn.neural_network import MLPClassifier
model=MLPClassifier(hidden_layer_sizes=(32,64,112), max_iter=5000)
```

```python
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report, accuracy_score
y_predict=model.predict(X_test)
```

```python
accuracy_score(y_test,y_predict)
```

```python
print(classification_report(y_test,y_predict))
```

```python
pd.crosstab(y_test,y_predict)
```
