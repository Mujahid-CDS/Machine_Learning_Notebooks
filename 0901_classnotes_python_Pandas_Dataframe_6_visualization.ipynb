# Drawing Multiple plots in same figure
# Subplot

```python
year=[2015,2016,2017,2018]
profit=[10, 15, 20, 12.5]

rcParams['figure.figsize'] = (7,7)

fig, ax=plt.subplots(2,2)

ax[0,0].plot(year,profit)
ax[0,0].set_title('First Plot')
```

```python
years = [2015, 2016, 2017, 2018]
profit = [10, 15, 20, 12.5]

rcParams['figure.figsize'] = (7,7)

fig, ax = plt.subplots(2,2)

ax[0,0].plot(years, profit)
ax[0,0].set_title('Apple')

ax[0,1].bar(years, profit)
ax[0,1].set_title('Banana')

ax[1,0].scatter(years, profit)
ax[1,0].set_title('Pineapple')

ax[1,1].pie(profit)
ax[1,1].set_title('Strawberry')

plt.show()
```

```python
years = [2015, 2016, 2017, 2018]
profit = [10, 15, 20, 12.5]

rcParams['figure.figsize'] = (6,3)

fig, ax = plt.subplots(1,2)

ax[0].bar(years, profit)
ax[0].set_title('Banana')

ax[1].scatter(years, profit)
ax[1].set_title('Pineapple')

plt.show()
```
